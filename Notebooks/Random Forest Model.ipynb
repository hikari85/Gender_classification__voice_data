```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import librosa

import warnings
from google.colab import drive
import os
import pandas as pd
```

Using MFCC Data

```python
mfcc_df = pd.read_csv("audio mfcc data.csv")

X = mfcc_df.iloc[:,:13]
y = mfcc_df.iloc[:,13]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)
```

```python
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=50)
rf_classifier.fit(X_train, y_train)

y_pred = rf_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

Using Chroma CENS Data

```python
chroma_df = pd.read_csv("audio chroma_cens data.csv")

X = chroma_df.iloc[:,:12]
y = chroma_df.iloc[:,12]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=10)
rf_classifier.fit(X_train, y_train)

y_pred = rf_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
