```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
import random

from glob import glob
import warnings

import librosa
import librosa.display
import IPython.display as ipd
from IPython.display import Audio

import os
import soundfile as sf
```

---

Final code to trim audios

```python
input_folder = "Audios/female/"

output_folder = "Audios/trim_audios/female/"

os.makedirs(output_folder, exist_ok=True)

for file_name in os.listdir(input_folder):
    input_file_path = os.path.join(input_folder, file_name)
    y, sr = librosa.load(input_file_path)

    start_time = 10
    end_time = 20
    start_sample = int(start_time * sr)
    end_sample = int(end_time * sr)

    trimmed_audio = y[start_sample:end_sample]

    output_file_path = os.path.join(output_folder, file_name)
    sf.write(output_file_path, trimmed_audio, sr)
```

```python
audio_file = "Audios/trim_audios/female/fe19.wav"
ipd.Audio(audio_file)
```
