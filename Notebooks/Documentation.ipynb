## 6. Machine Learning Models

The following machine learning models were employed for predicting gender based on audio:

1. **Logistic Regression:** A statistical model for binary classification (e.g., male vs. female). It calculates the probability of a class and classifies based on a threshold.

2. **CNN (Convolutional Neural Network)**: A deep learning model that processes audio spectrograms, extracting complex patterns and features to classify gender. CNNs are good at handling large, high-dimensional data like audio.

3. **Random Forest**: An ensemble learning method using multiple decision trees to make predictions. It combines the outputs of many trees for robust and accurate classification, handling complex and noisy data well.

4. **Decision Tree:** A model that uses a tree-like structure for classification based on splitting data based on feature values. It is easy to interpret and handles both categorical and continuous data, but can be prone to overfitting.

## 7. Features Used for Training
* MFCC (Mel-frequency cepstral coefficients) - 13 features

  * Definition: MFCCs are a set of features that represent the short-term power spectrum of a sound signal, designed to mimic the human auditory system's response.
  * Derivation Steps:

    * Fourier Transform:
    Apply a Fourier transform to a windowed excerpt of the signal to obtain its power spectrum.
    * Mel Scale Mapping:
    Map the power spectrum to the mel scale using triangular or cosine overlapping windows.
    * Log Transformation:
    Compute the logarithm of the powers at each mel frequency.
    * Discrete Cosine Transform (DCT):
    Perform a DCT on the mel log powers.
    * Extract MFCCs:
    The resulting amplitudes from the DCT are the MFCCs.

* Chroma Cens (Chroma Energy Normalized Statistics)- 12 features

 * Defination: Chroma Cens is a feature extraction method widely used in music information retrieval.

  * It refines chroma features, which represent the distribution of musical pitches in an audio signal, by normalizing them with respect to energy.

  * This normalization enhances their robustness to variations in overall loudness. Chroma CENS then computes statistical descriptors, such as mean and standard deviation, over short time frames, capturing the temporal evolution of chroma distributions.

  * These features provide a compact yet informative representation of the harmonic content of music signals, making them valuable for tasks like music genre classification, chord recognition, and similarityÂ analysis.
